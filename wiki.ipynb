```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
url = 'https://leagueoflegends.fandom.com/wiki/List_of_champions'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc, features='html.parser')
```

```python
print("ราชื่อ Champions ใน League of Legends")
print('-'*80)
champs = []

rows = soup.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    if len(cols) > 0:
        champion_name = cols[0].text.strip()
        champ_class = row.find("a", class_="mw-redirect")
        if champ_class:
            champion_class = champ_class.text.strip()
            champs.append((champion_name, champion_class))
    


for champion in champs[2:]:
    print(champion[0] + " มีคลาสเป็น " + champion[1])
print(len(champs))
```

```python
status_page = dict()
for i in range(1, 494):
    url = "https://www.mobafire.com/league-of-legends/blog?page=" + str(i)
    print("Collecting from {}".format(url))
    result = requests.get(url)
    time.sleep(5)

    html_doc = result.text
    soup = BeautifulSoup(html_doc, features="html.parser")

    # Extract data from HTML
    content = soup.find_all("article", {"class": "box-shadow-lb"})

    for row in content:
        title = row.find("a", {"href": True}).text
        date = row.find("span", {"class": "localized-datetime tooltip noTouchTip"}).text
        status_page[date]=title
```

```python
status = pd.DataFrame({"Date":status_page.keys(),"Title":status_page.values()})
status.set_index("Date",inplace = True)
status
```

```python
status.to_csv('mobafire.csv')
```
