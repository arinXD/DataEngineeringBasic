```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
```

```python
dToday = datetime.today().strftime('%Y-%m-%d')
```

```python
url = 'https://leagueoflegends.fandom.com/wiki/List_of_champions'
r = requests.get(url)
html_doc = r.text
soup = BeautifulSoup(html_doc, features='html.parser')
```

```python
print("ราชื่อ Champions ใน League of Legends")
print('-'*100)
champions = []

# soup.find("div",{"id":"pt-topic-left"})
table = soup.find('table', {"class":"article-table"})
tbody = table.find("tbody")
tr = tbody.find_all('tr')
for i in tr:

    span = i.find('span', attrs={'style':'white-space:normal;'})
    if(span.a is None): continue

    champoinClass = i.find_all('a', {'class':{'mw-redirect'}})

    className = []
    champoinName = span.a.contents[0]
    
    if(champoinClass!=[]):
        for j in champoinClass:
            className.append(j.text)
    else:
        a = i.find('a', attrs={'title':'Champion classes/Tank'})
        className.append(a.text)
    print([champoinName, ", ".join(className)])
    champions.append([champoinName, ", ".join(className)])    

# for row in rows:
#     cols = row.find_all('td')
#     if len(cols) > 0:
#         champion_name = cols[0].text.strip()
#         champ_class = row.find("a", class_="mw-redirect")
#         if champ_class:
#             champion_class = champ_class.text.strip()
#             champs.append((champion_name, champion_class))

# for champion in champs[2:]:
#     print(champion[0] + " มีคลาสเป็น " + champion[1])
# print(len(champs))
df = pd.DataFrame(champions, columns=['Champion Name','Classes'])
df.to_csv("leagueoflegends_champions_{}.csv".format(dToday))
```
